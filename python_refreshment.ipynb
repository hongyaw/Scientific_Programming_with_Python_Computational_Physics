Save each line one after another

```python
file = open('storelist3.dat','w')
for i in range(1,6):
    file.write(str(i**2))
    file.write("\n")
file.close()
```

```python
with open('storelist3.dat','w') as file:
    for i in range(1,6):
        file.write(str(i**2))
        file.write("\n")  ##Dont even need to close the file as it happens in the block only
```

### 3d plots

```python
x2,y2=np.meshgrid(
    np.linspace(-10,10,201),
    np.linspace(-10,10,201)
)
z2 = x2 + y2**2
```

```python
contourplot = plt.contour(x2,y2,z2)
plt.clabel(contourplot, inline = 1, fontsize = 10)
```

```python
x3,y3=np.meshgrid(
    np.linspace(-10,10,201),
    np.linspace(-10,10,201)
)
z3 = np.cos(x3+y3) + 0.05*(x3-y3)

plt.contourf(x3,y3,z3)
plt.colorbar()
```

```python
plotproj = plt.axes(projection='3d')
plotproj.view_init(70,30)
plotproj.contour3D(x3,y3,z3,100,cmap='binary')
```

### 3d plots - lines & scatter plot

```python
plt3D = plt.axes(projection='3d')
z1 = np.linspace(0,30,301)
x1 = np.sin(z1)
y1 = np.cos(z1)
plt.xlim([-2,2])
plt.ylim([-2,2])
plt3D.plot3D(x1,y1,z1)
```

```python
plt3D = plt.axes(projection='3d')
z1 = np.linspace(0,10,51)
x1 = np.sin(z1)
y1 = np.cos(z1)
plt.xlim([-2,2])
plt.ylim([-2,2])
plt3D.scatter3D(x1,y1,z1)
```
